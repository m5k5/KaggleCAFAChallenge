```python
import tensorflow as tf
import numpy as np
```

```python
from tensorflow.keras import layers


def createModel():
    inputs = tf.keras.Input(shape=(5,))
  
    x = layers.Dense(32)(inputs)
    x = layers.LeakyReLU()(x)
    x = layers.Dense(16)(x)
    x = layers.LeakyReLU()(x)
    x = layers.Dense(32)(x)
    x = layers.LeakyReLU()(x)

    outputs=layers.Dense(5, activation=tf.keras.activations.sigmoid)(x)

    return tf.keras.Model(inputs=inputs, outputs=outputs, name="DenseModel")

model = createModel()

model.summary()
```

```python
rand=np.random.randint(2, size=(10,5))
X=rand
y=rand
```

```python
optim = tf.keras.optimizers.Adam()
loss = tf.keras.losses.BinaryCrossentropy()
acc = tf.keras.metrics.BinaryAccuracy()
prec = tf.keras.metrics.Precision()
rec = tf.keras.metrics.Recall()
model.compile(optim, loss, metrics=[acc,prec,rec])
model.fit(X,y, epochs=200)
```
