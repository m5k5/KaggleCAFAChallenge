```python
import matplotlib.pyplot as plt
import pandas as pd
import os

DATA_PATH = os.getenv('DATA_PATH')
print(DATA_PATH)
```

```python
df = pd.read_csv(os.path.join(DATA_PATH, "trainingHistory", "DenseModelIpAndBioProps_20230801-164846_CCO.csv"))
DenseIPF1 = df["Value"].to_numpy()

df = pd.read_csv(os.path.join(DATA_PATH, "trainingHistory", "DenseModelIpAndBioProps_20230801-164846_CCO val.csv"))
DenseIPF1Val = df["Value"].to_numpy()

df = pd.read_csv(os.path.join(DATA_PATH, "trainingHistory", "DenseMultiModal_20230808-220442_CCO.csv"))
MultiModal1F1 = df["Value"].to_numpy()

df = pd.read_csv(os.path.join(DATA_PATH, "trainingHistory", "DenseMultiModal_20230808-220442_CCO val.csv"))
MultiModal1F1Val = df["Value"].to_numpy()

df = pd.read_csv(os.path.join(DATA_PATH, "trainingHistory", "DenseMultiModal_20230808-235835_CCO.csv"))
MultiModal2F1 = df["Value"].to_numpy()

df = pd.read_csv(os.path.join(DATA_PATH, "trainingHistory", "DenseMultiModal_20230808-235835_CCO val.csv"))
MultiModal2F1Val = df["Value"].to_numpy()
```

```python
from scipy.ndimage import gaussian_filter1d

plt.figure(figsize=(7,12))

plt.subplot(2,1,1)
_ = plt.plot(gaussian_filter1d(DenseIPF1, sigma=3), label="InterPro Model Run 1")
_ = plt.plot(gaussian_filter1d(MultiModal1F1, sigma=3), label="InterPro Model Run 2")
_ = plt.plot(gaussian_filter1d(MultiModal2F1, sigma=3), label="Multimodal Model")
_ = plt.xlabel("steps")
_ = plt.ylabel("Training F1")
_ = plt.legend()
_ = plt.title("Training")

plt.subplot(2,1,2)
_ = plt.plot(gaussian_filter1d(DenseIPF1Val, sigma=1), label="InterPro Model Run 1")
_ = plt.plot(gaussian_filter1d(MultiModal1F1Val, sigma=1), label="InterPro Model Run 2")
_ = plt.plot(gaussian_filter1d(MultiModal2F1Val, sigma=1), label="Multimodal Model")
_ = plt.xlabel("epochs")
_ = plt.ylabel(" Validation F1")
_ = plt.legend()
_ = plt.title("Validation")

# plt.tight_layout()
```
